```python
def extract_registration_details(row):
    month, year = np.nan, np.nan
    if row is not np.nan:
        if ' ' in row:
            parts = row.split(' ')
            month = parts[0] if parts[0].isalpha() else np.nan
            year = parts[1] if parts[1].isdecimal() else np.nan
        elif row.isalpha():
            month = row
            year = np.nan
        else :
            month = np.nan
            year = row
    return month, year

df[['Reg_mnt', 'Reg_year']] = df['Reg_yr'].apply(extract_registration_details).apply(pd.Series)
```

```python
def extract_manufacturing_details(row):
    month, year = np.nan, np.nan
    if row is not np.nan:
        if ' ' in row:
            parts = row.split(' ')
            month = parts[0] if parts[0].isalpha() else np.nan
            year = parts[1] if parts[1].isdecimal() else np.nan
        elif '-' in row:
            year, month = row.split('-')[0], row.split('-')[1]
        elif row.isdecimal() :
            month = np.nan
            year = row
        else :
            month = row
            year = np.nan
            
    return month, year

df[['Make_mnt','Make_yr']] = df['Manufacturing_yr'].apply(extract_manufacturing_details).apply(pd.Series)
```

```python
def km_extractor(row):
    kms = np.nan, 
    if pd.notna(row):
        if 'km' in row:
            kms = row.split('km')[0]
        elif 'Kms' in row:
            kms = row.split('Kms')[0]
        if ',' in kms:
            kms = kms.replace(',','')
        elif 'K' in kms:
            kms = kms.replace('K', '000')
    return kms

df['Km_driven'] = df['Km_driven'].apply(km_extractor)
```

```python
def extract_owner_no(row):
    if pd.notna(row):
        if 'first' in row.lower() or '1' in row.lower() :
            return '1'
        elif 'second' in row.lower() or '2' in row.lower() :
            return '2'
        elif 'third' in row.lower() or '3' in row.lower() :
            return '3'   
        elif 'fourth' in row.lower() or '4' in row.lower() :
            return '4'  
        elif 'fifth' in row.lower() or '5' in row.lower() :
            return '5' 
        else:
            return '0'   
    return row

df['Prev_owners'] = df['Prev_owners'].apply(extract_owner_no)
```

```python
def normalize_fueltype(ftype):
    if pd.notna(ftype):
        ftype = ftype.lower()
        if '+' in ftype or 'hybrid' in ftype:
            return 'Hybrid'
        elif 'petrol' in ftype:
            return 'Petrol'
        elif 'diesel' in ftype:
            return 'Diesel'
        elif 'cng' in ftype:
            return 'CNG'
        elif 'electric' in ftype:
            return 'Electric'
        elif 'lpg' in ftype:
            return 'LPG'
        else:
            return 'Other'
    return ftype

df['Fueltype'] = df['Fueltype'].apply(normalize_fueltype)
```

```python
def clean_rto(row):
    if pd.notna(row):
        for char in row:
            if char.isdigit():
                return row
        row = row.lower()
        if row == 'not available':
            return np.nan
        return row.lower()
    return np.nan

df['RTO'] = df['RTO'].apply(clean_rto)
```

```python
def clean_rto_code(row, mapper=rto_state_mapping):
    if pd.notna(row):
        if row in mapper.keys():
            return mapper[row] + 'XX'
        else :
            return row
    return np.nan

df['RTO_code'] = df['RTO'].apply(clean_rto_code)
```

```python
def extract_rto_city_n_state(row):
        if pd.notna(row):
            if '*' in row:
                row = row.split('*')[0]
            if row.strip() == '':
                return np.nan
            elif row == row.upper() and row[:2].isalpha():
                return row[:2]
        return np.nan

df['RTO_state'] = df['RTO_code'].apply(extract_rto_city_n_state)
```

```python
def rto_city_code(row):
    if pd.notna(row):
        if len(row) <= 2:
            return np.nan
        if len(row) > 3 and row[2:4].isdecimal():
            return row[2:4]
        elif row[2].isdecimal():
            return '0'+row[2]
        else :
            return np.nan
    return row

df['RTO_city'] = df['RTO_code'].apply(rto_city_code)
df['RTO_city']
```

```python
import re

def clean_maxpower(row):
   if pd.notna(row) :
      row = row.lower()
      if 'doctype' in row:
         return row.split('@')[0], np.nan
      elif '@' in row:
         bhp, rpm = row.split('@')[0],row.split('@')[1]
         bhp, rpm = re.sub(r'\D', '', bhp).strip(), re.sub(r'\D', '', rpm).strip()
      elif 'at' in row:
         bhp, rpm = row.split('ps at')[0],row.split('ps at')[1]
         bhp, rpm = re.sub(r'\D', '', bhp).strip(), re.sub(r'\D', '', rpm).strip()
      elif '-' in row:
         bhp, rpm = row.split('-')[0],row.split('-')[1]
         bhp, rpm = re.sub(r'\D', '', bhp).strip(), re.sub(r'\D', '', rpm).strip()         
      elif 'bhp' in row and 'rpm' not in row:
         bhp,rpm = re.sub(r'\D', '', row).strip(), np.nan
      elif 'bhp' not in row and 'rpm' in row:
         bhp,rpm =  np.nan, re.sub(r'\D', '', row).strip()
      else :
         bhp, rpm = np.nan, np.nan
      if bhp == '':
         bhp = np.nan
      if rpm == '':
         rpm = np.nan 
      return bhp, rpm
   return np.nan, np.nan
df[['Max bhp', 'Max rpm']] = df['Max Power'].apply(clean_maxpower).apply(pd.Series)
```
